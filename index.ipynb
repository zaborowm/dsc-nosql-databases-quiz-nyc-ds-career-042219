## Question 1

What is the difference between a NoSQL and a SQL database?  Which is better?

_______________________________________________________________________________________________________________________________

SQL database is structured, relational, requires definition of fields up front - predefined schemas. They can handle large volumes, but if any new data types are required, the db needs to be redesigned from scratch, thus are not flexible to changing environments. Strength is the well-defined, structured data, not as effecive with unstructured data.

NoSQL database is more document-oriented and distributed, can add data types over time without needing to adjust the schema - have dynamic schema - making them very effective for unstructured data.

Better for different things.  NoSql better for big data, text or document. SQL better for activities requiring consistent, non-changing fields, such as in operations.

## Question 3

How does MongoDB work? How is it different from a relational database?

Write your answer below this line: 
_______________________________________________________________________________________________________________________________

Mongo DB is an opensource NoSQL document database, designed to handle large volumes of unstructured data through a document metaphor. It uses field-value pairs (like a dictionary) to store data, and thus is flexible as to the type of data that can be stored as well as the type of data. A relational database uses structured tables and schema to store its data. 

Mongo can be more flexible in deployment, as it operates on commodity hardware and is horizontally scaleable - can use in cloud or on premise and if you need more power, space, growth you can add more servers. SQL databases require reconfiguration to different physical architectures, which is sensitive as the database integrity depends on the structure.

Mongo allows for faster feture development, as you need new capabiity, fields can be added to the documents. In SQL, you need to redesign the schema and reestablish the full database structure prior to initiating development changes.






## Question 4

What is fault tolerance, and what does it have to do with Resilient Distributed Datasets?

Write your answer below this line: 
_______________________________________________________________________________________________________________________________

Fault tolerance is the ability to maintain operation when a server or component is down.  Resilient distributed datasets are when data is stored redundantly across servers, in the event that the primary sources are not avaiable due to a hardware failure or network outage. Resilient distributed datasets can maintain operations by referring to the backups - uninterrupted operations.
